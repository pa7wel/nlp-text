```python
import spacy
from spacy import displacy

# Load English tokenizer, tagger, parser and NER
nlp = spacy.load("en_core_web_sm")
```

```python
with open ('data/wiki_us.txt', 'r') as f:
    text = f.read()
with open ('data/article1.txt', 'r') as a:
    article = a.read()
```

```python
doc = nlp(text)
doc_article = nlp(article)
print(doc)
```

```python
print (len(doc))
print (len(text))
for token in doc[:10]:
    print (token)
```

```python
sentence1 = list(doc.sents)[0]
print (sentence1)
```

```python
token2 = sentence1[2]
print (token2)
token2.text
token2.left_edge
token2.right_edge
print (token2.ent_type_)
```

```python
# NER Named Entity Recognition
for ent in doc.ents:
    print (ent.text, ent.label_)
```

```python

displacy.render(doc, style='ent', jupyter=True)
```

```python
# NER Named Entity Recognition
for ent in doc_article.ents:
    print (ent.text, ent.label_)

displacy.render(doc_article, style='ent', jupyter=True)
```
